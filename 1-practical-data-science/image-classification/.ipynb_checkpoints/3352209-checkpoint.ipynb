## Part 1 | Data Processing
---

As a first step we download all the needed data for this exercise. A `data/` folder was created under the project's directory where the two image datasets where stored into separate sub-folders, as well the Excel info file for the `LREC2022` dataset. The following cells use an image reading function to create a dataframe for every different image dataset. The function reads an image, converts it to the grayscale color space and resizes it to a `28x28` format. Each dataframe consists of the file name of the image, the character that the image displays, the century of the character and two image columns, one that corresponds to the original format and one flattend. Especially for the `LREC2022` dataset, we remove images that correspond to the `NaN` character.

It is worth noting that the sub-folders of the image datasets contained the `.DS_Store` hidden file. The file was removed from the overall directory with the use of the following [bash command](https://bit.ly/3GO6s9x): `find . -name \*.DS_Store -type f -delete`.

```python
def read_image(path: str, h: int=28, w: int=28) -> list:

    """
    Read and return a (h, w) array of the image from given directory.

    PARAMETERS
        @ path: string of directory
        @ h: height of desired image resize
        @ w: width of desired image resize
        @ color: color space conversion code

    RETURNS
        (h, w) array of image
    """

    img = io.imread(path)
    img = resize(img, (h, w), anti_aliasing=True)
    img = color.rgb2gray(img[:, :, :3]) # discard alpha (transparency) channel, function expect a 3-dimension RGB image
    return img
```

```python
# LREC Dataset

info = pd.read_excel('./data/OCs10-12CE.xlsx') # read Excel file

names, images = [], []
data_dir = './data/LREC2022_image_characters/'
for img_path in os.listdir(data_dir):
    curr = read_image(f"{data_dir}/{img_path}")
    images.append(curr) # image
    names.append(img_path) # name of image file

ds1 = pd.DataFrame({
    'name': names,
    'image': images
})
ds1['flattend'] = ds1['image'].apply(lambda x: x.flatten())
ds1 = pd.merge(info, ds1, left_on='file name', right_on='name') # merge Excel and image DataFrames on same file names
ds1 = ds1[['name', 'character', 'century', 'image', 'flattend']]
ds1 = ds1.dropna().reset_index(drop=True) # drop nan character
ds1.head()
```

```python
# BT1 Dataset

names, characters, images, tm = [], [], [], []
data_dir = './data/BT1_BigByLetters2022'
for img_subdir in os.listdir(data_dir): # traverse symbol sub-folders
    for img_path in os.listdir(f"{data_dir}/{img_subdir}"): # traverse images of symbol
        curr = read_image(f"{data_dir}/{img_subdir}/{img_path}")
        images.append(curr) # image
        names.append(img_path) # name of image file
        characters.append(img_subdir) # character symbol
        tm.append(img_path.split('_')[1]) # TM

ds2 = pd.DataFrame({
    'name': names,
    'character': characters,
    'tm': tm,
    'image': images,
})
ds2['flattend'] = ds2['image'].apply(lambda x: x.flatten())
ds2.head()
```

For this part we will visualize the character and century frequency for the two datasets. We will use the following function that plots a frequency bar plot of the given dataframe and column.

```python
def plot_value_counts(df: pd.DataFrame, column: str, title: str, figsize: Tuple[int, int]=(14, 6), offset: int=5, reset_index: bool=False, horizontal: bool=False) -> None:

    """
    Plots a value_counts bar plot of the given dataframe's column with respective data labels.

    PARAMETERS
        @ df: pandas DataFrame
        @ column: string, column of dataframe
        @ title: string, title of figure
        @ figsize: tuple, figure size (width, height)
        @ offset: int, data annotation placement offset
        @ reset_index: bool, reset_index in case value_counts are not string components
        @ horizontal: boolean, if True plot a horizontal bar plot

    RETURNS
        None
    """

    current = df[column].value_counts().sort_index() # sorting enables correct label annotation

    orient = 'bar'
    if horizontal:
        orient = 'barh'

    plot = current.plot(kind=orient, title=title, rot=0, color=['tab:blue'], figsize=figsize)

    i = 0
    # https://bit.ly/3DteCBO | place value annotations at the center of the bars
    for rect in plot.patches:
        x_value = rect.get_x() + rect.get_width() / 2
        y_value = rect.get_height() + offset
        if horizontal:
            x_value = rect.get_width() + offset
            y_value = rect.get_y()
        if reset_index:
            plt.annotate(current.reset_index(drop=True)[i], xy = (x_value, y_value), ha='center', fontsize=10, color='tab:orange')
        else:
            plt.annotate(current[i], xy = (x_value, y_value), ha='center', fontsize=10, color='tab:orange')
        i += 1
```

## Part 3.1 | Image Classification with respect to Character
---

The machine learning approach we are going to follow in order to classify the characters of the two datasets are Convolutional Neural Networks. The decision is mainly based on recent scientific research ([1](https://bit.ly/3VxPIqY), [2](https://bit.ly/2OUuhSd), [3](https://bit.ly/3XAPrW2)) proving that CNNs are effective at tackling image-related problems. Since we are using this method, we are not going to apply any filters to the images prior feeding them to the training model. This is due to the fact that the convolutional layers of the network will take care of feature identification between the characters and hence, in the writer's opinion, there is no reason risking potential decrease in the quality of the image by applying contrast enhancement filters or edge detection algorithms.

In order to handle the imbalance we discovered in the previous step, we will compute the class weight of every character and every dataset. The resulting dictionary will be passed to the training step of the network, indicating to the algorithm from which class the model ought to learn more. We note that instead of the flattend arrays, the original array format of the image will be used, since CNNs accept such inputs. Last but not least, we will convert the labels, that is the characters, to categorical entities before passing them to the training model. We note that our labels are not one-hot-encoded, but integers from 0 up to 28.

```python
# Prepare data for CNN input
x_ds1 = np.array(ds1['image'].tolist()) # select just the 2d images
_, y_ds1 = np.unique(ds1['character'], return_inverse=True) # https://bit.ly/3ATSSNN | convert numerical classes to categorical

x_ds2 = np.array(ds2['image'].tolist())
_, y_ds2 = np.unique(ds2['character'], return_inverse=True)

# Compute class weight
class_weight_ds1 = compute_class_weight(class_weight='balanced', classes=np.unique(y_ds1), y=y_ds1)
class_weight_ds1 = dict(enumerate(class_weight_ds1)) # https://bit.ly/3EU5NAL | numpy array to dictionary
class_weight_ds2 = compute_class_weight(class_weight='balanced', classes=np.unique(y_ds2), y=y_ds2)
class_weight_ds2 = dict(enumerate(class_weight_ds2))
```

For the BT1 dataset we are going to split the data in 80% training and 20% testing. From the training data we will use 15% for validation. For the LREC2022 dataset, we will increase the size of the test set to 45% and keep the validation percentage intact. The reason for that increase is to enable all different classes to be present at the train and test subsets. Of course, the risk here is that our validation set will not contain all 28 classes, which is not necessarily bad. However, we'll have to remember that detail in case over-fitting effects occur in the learning curve visualizations, since these might not correspond to actual over-training.

```python
# Dataset 1
X_train_ds1, X_test_ds1, y_train_ds1, y_test_ds1 = train_test_split(x_ds1, y_ds1, test_size=0.45, random_state=39) # split test
X_train_ds1, X_val_ds1, y_train_ds1, y_val_ds1 = train_test_split(X_train_ds1, y_train_ds1, test_size=0.15, random_state=39) # split train, validation

# Dataset 2
X_train_ds2, X_test_ds2, y_train_ds2, y_test_ds2 = train_test_split(x_ds2, y_ds2, test_size=0.20, random_state=39) # split test
X_train_ds2, X_val_ds2, y_train_ds2, y_val_ds2 = train_test_split(X_train_ds2, y_train_ds2, test_size=0.15, random_state=39) # split train, validation

# Summary
print('Dataset 1')
print('Train:       X=%s, y=%s' % (X_train_ds1.shape, y_train_ds1.shape))
print('Validation:  X=%s,  y=%s' % (X_val_ds1.shape, y_val_ds1.shape))
print('Test:        X=%s, y=%s' % (X_test_ds1.shape, y_test_ds1.shape))
print('----------------------------------------')
print('Dataset 2')
print('Train:       X=%s, y=%s' % (X_train_ds2.shape, y_train_ds2.shape))
print('Validation:  X=%s,  y=%s' % (X_val_ds2.shape, y_val_ds2.shape))
print('Test:        X=%s, y=%s' % (X_test_ds2.shape, y_test_ds2.shape))
```

For the training procedure we will use the `input_fn` function, which basically receives X and y training data and returns `PrefetchedDataset` objects. The latter are Tensorflow objects that enable faster training, since the model can read the next-to-train batch of data while training the current batch. The batch size, which is set to 32, correspond to the number of images that will flow through the network before a back-propagation operation, and hence a fine-tuning of the weights, takes place. While batch sizes can [significantly alter](https://bit.ly/3OSe50t) a networks' behavior, we are not going to be testing different batch sizes.

```python
BATCH_SIZE = 32
AUTOTUNE = tf.data.experimental.AUTOTUNE # size of prefetched data will be optimized at runtime

def input_fn(X: np.ndarray, y: np.ndarray, batch_size: int=BATCH_SIZE):

    """
    Receives X and y training data and transforms them in PrefetchDataset objects.
    The data is shuffled and batched in the specified size.
    tf.data.experimental.AUTOTUNE is used for calculating the size of prefetched data at runtime according to resource heuristics.

    PARAMETERS
        @ X: X training numpy array
        @ y: y training numpy array
        @ BATCH_SIZE: int corresponding to the training batch size

    RETURNS
        PrefetchedDataset object dataset
    """

     # https://bit.ly/3uiStRi | shuffle, repeat, batch
    ds = tf.data.Dataset.from_tensor_slices((X, y)) # initialize dataset
    ds = ds.shuffle(buffer_size=len(X_train_ds1)) # shuffle data, buffer size = maximum size of training data
    ds = ds.repeat() # repeat indefinitely
    ds = ds.batch(batch_size) # batch in specified size
    ds = ds.prefetch(buffer_size=AUTOTUNE) # PrefetchDataset object
    return ds
```

```python
# Dataset 1
train_ds1 = input_fn(X_train_ds1, y_train_ds1, BATCH_SIZE) # PrefetchedDataset objects
validation_ds1 = input_fn(X_val_ds1, y_val_ds1, BATCH_SIZE)
test_ds1 = input_fn(X_test_ds1, y_test_ds1, BATCH_SIZE)

# Dataset 2
train_ds2 = input_fn(X_train_ds2, y_train_ds2, BATCH_SIZE) # PrefetchedDataset objects
validation_ds2 = input_fn(X_val_ds2, y_val_ds2, BATCH_SIZE)
test_ds2 = input_fn(X_test_ds2, y_test_ds2, BATCH_SIZE)
```

The model we are going to develop is inspired by [LeNet](https://bit.ly/2oMP8rP), one of the earliest CNN architectures proposed to solve character recognition problems. The optimizer we chose for our model is Adam, which according to [recent scientific research](https://bit.ly/3Fhpp3h) seems to have an edge on character recognition tasks with respect to other optimization techniques. To avoid over-fitting, dropout layers were added to the model's architecture, which randomly ignore a subset of layer outputs. The loss function the training model is trying to minimize is the Sparse Categorical Cross Entropy, since our labels are not one-hot-encoded. We note that the last fully-connected layer of the network is set to the number of classes of the classification task. That means that the network will have 24 and 28 nodes for datasets 1 and 2, respectively.

```python
def init_adam_model(summary: bool, nodes: int) -> models.Sequential:

    """
    Sets up and returns a LeNet inspired CNN architecture.
    Optimizer: Adam
    Loss: Sparse Categorical Cross Entropy

    PARAMETERS
        @ summary: boolean, display model summary
        @ nodes: number of nodes at the last fully-connected layer

    RETURNS
        keras sequential model
    """

    # https://bit.ly/3Ummevd | LeNet architecture
    model = models.Sequential()

    model.add(layers.Conv2D(20, 5, activation='relu', padding='same', input_shape=(28, 28, 1)))
    model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    model.add(layers.Dropout(0.10))

    model.add(layers.Conv2D(50, 5, activation='relu', padding="same"))
    model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    model.add(layers.Dropout(0.20))

    model.add(layers.Flatten())

    model.add(layers.Dense(500, activation='relu'))

    model.add(layers.Dense(nodes, activation='softmax'))

    model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001), loss=tf.keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])

    if summary:
        model.summary()

    return model
```

Last but not least we will implement the following functions that help us visualize the learning curves of the training process and evaluates the model on the test set:

```python
def summarize_diagnostics(history: tf.keras.callbacks.History) -> plt.figure :

    """
    Plots training and validation learning curves.
    Two subplots are created, each one of them comparing training/validation loss and accuracy respectively.

    PARAMETERS
        @ history: model history callback

    RETURNS
        a matplot figure consisting of two subplots, one for loss, one for accuracy

    """

    plt.figure(figsize=(16, 8))
    plt.suptitle('Training Curves')
    plt.subplot(1, 2, 1) # plot loss
    plt.title('Cross Entropy Loss')
    plt.plot(history['loss'], color='tab:blue', label='train')
    plt.plot(history['val_loss'], color='tab:orange', label='val')
    plt.legend(loc='upper right')
    plt.subplot(1, 2, 2) # plot accuracy
    plt.title('Classification Accuracy')
    plt.plot(history['accuracy'], color='tab:blue', label='train')
    plt.plot(history['val_accuracy'], color='tab:orange', label='val')
    plt.legend(loc='lower right')
    return plt

def model_evaluation(model: models.Sequential, test_ds, X_test: np.ndarray) -> None:

    """
    Evaluates given model to given test set.
    Prints test and evaluation metrics.

    PARAMETERS
        @ model: keras model that will be evaluated on given testing data
        @ test_ds: PrefetchedDataset object corresponding to testing data X, y
        @ X_test: numpy array corresponding to X testing data, used for calculating evaluations steps

    RETURNS
        None
    """

    print('Test set evaluation metrics')
    loss, accuracy = model.evaluate(test_ds, steps=X_test.shape[0]//BATCH_SIZE) # steps are given since our data is PrefetcedDataset objects!
    print("Cross Entropy Loss: {:.2f}".format(loss))
    print("Accuracy: {:.2f}".format(accuracy))

def model_report(model: models.Sequential, history: tf.keras.callbacks.History, test_ds, X_test: np.ndarray) -> None:

    """
    Wrapper function of summarize_diagnostics and model_evaluation functions.
    Evaluates given model on testing data.
    Plots learning curves for training and validation steps.

    PARAMS
        @ model: keras model that will be evaluated on given testing data
        @ history: model history callback
        @ test_ds: PrefetchedDataset object corresponding to testing data X, y
        @ X_test: numpy array corresponding to X testing data, used for calculating evaluations steps

    RETURNS
        None
    """

    plt = summarize_diagnostics(history)
    plt.show()
    model_evaluation(model, test_ds, X_test)
```

### Dataset 1 | LREC2022

We will start training for the first dataset:

```python
model_ds1 = init_adam_model(summary=True, nodes=28) # 28 classes for LREC2022 dataset
```

We can see that our model has about 1 million trainable parameters. We move on to training the model for 50 epochs:

```python
model_train_ds1 = model_ds1.fit(
    train_ds1,
    epochs=50,
    steps_per_epoch=X_train_ds1.shape[0]//BATCH_SIZE,
    validation_data=validation_ds1,
    validation_steps=X_val_ds1.shape[0]//BATCH_SIZE,
    class_weight=class_weight_ds1
)
```

```python
model_report(model_ds1, model_train_ds1.history, test_ds1, X_test_ds1)
```

We notice that even though the learning curves indicate over-fitting, the testing accuracy complies with the validation one. This was expected, as mentioned, to happen, since our validation set was not consisting of all the available 28 classes. We note the classification accuracy of 61%, and we move on the respective confusion matrix of the model:

```python
predict_ds1 = model_ds1.predict(X_test_ds1) # prediction probabilities for every class
y_predicted_ds1 = np.argmax(predict_ds1, axis=1) # https://bit.ly/3Vebalb | calculate max probability for every row = predicted class

conf_matrix = confusion_matrix(y_test_ds1, y_predicted_ds1)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_matrix, display_labels=np.unique(ds1['character']))
fig, ax = plt.subplots(figsize=(8, 8))
cm_display.plot(ax=ax, cmap='BuGn')
ax.set_title('Confusion Matrix LREC2022 Dataset | Character Classification')
ax.set_xlabel('predicted')
ax.set_ylabel('true')
plt.show()
```

```python
print(classification_report(y_test_ds1, y_predicted_ds1, zero_division=0, target_names=np.unique(ds1['character']).tolist()))
```

The diagonal of the matrix is relatively distinguishable in the model's confusion matrix. However, the small size of our data is noticeable deeming our classifier ambiguous. Lastly, we note that even though there are quite a lot miss-classifications taking place, these are happenning in low numbers for each character. In any case, characters `ε`, `κ`, and `τ` are most commonly miss-classified as `σ`, `η` and `ι`, respectively. In regards to the classification report metrics, we note that `λ` achieves the highest precision among 9 instances, meaning that whenever the model classified a character as `λ`, it was 100% correct. The highest recall is computed for `ν`, which means that out of the total `ν` image instances that were given to the model, it was able to successfully identify 90% of them. Last but not least, the highest F1-score of the report is equal to 81% and is appointed to `o`.

## Part 3.2 | Image Classification with respect to Century
---

We will now classify the images of the two datasets again, but this time we'll use the century and TM columns as our labels. The procedure for training the model will be exactly the same. For starters, we will update our training, validation and test datasets:

## Part 4 | Supervised KMeans Clustering on LREC2022 Dataset
---

We will cluster the first dataset of images using the character and century as the ground truth. First things first, we alter our training datasets accordingly:

```python
x_ds1 = np.array(ds1['flattend'].tolist())
y_ds1_character = np.array(ds1['character'].tolist())
y_ds1_century = np.array(ds1['century'].tolist())
```

We will split our dataset to train and test subsets:

```python
X_train_ds1, X_test_ds1, y_train_ds1, y_test_ds1 = train_test_split(x_ds1, y_ds1_character, test_size=0.20, random_state=39)  # split train/test
```

We will utilize the Normalised Mutual Information metric in order to identify the optimal k for clustering our images. We won't test k values higher than 30, since our characters are only 28:

```python
opt_k = 0
optimal = 0
for k in range(1, 30):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=39) # init KMeans
    kmeans.fit(X_train_ds1) # fit to training data
    y_predict_ds1 = kmeans.predict(X_test_ds1) # predict on test
    curr = normalized_mutual_info_score(y_test_ds1, y_predict_ds1) # compute metric
    if curr > optimal: # maximize Normalised Mutual Information
        optimal = curr
        opt_k = k

print("Optimal k: {}, NMI: {}".format(opt_k ,optimal))
```

The optimal k is found to be 28, which is expected considering the fact that our dataset indeed consists of 28 different classes.

We move on the clustering with respect to century. We will first split our data once again into training and test:

```python
X_train_ds1, X_test_ds1, y_train_ds1, y_test_ds1 = train_test_split(x_ds1, y_ds1_century, test_size=0.20, random_state=39)  # split train/test
```

We set the loop to run up to 10, taking due consifderation the 3 centuries that our data consists of:

```python
opt_k = 0
optimal = 0
for k in range(1, 10):
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=39)
    kmeans.fit(X_train_ds1)
    y_predict_ds1 = kmeans.predict(X_test_ds1)
    curr = normalized_mutual_info_score(y_test_ds1, y_predict_ds1)
    if curr > optimal:
        optimal = curr
        opt_k = k

print("Optimal k: {}, NMI: {}".format(opt_k ,optimal))
```

The optimal k equals to 2 and once again falls around the expected classes, which is 3. As described in the classification task, a behaviour like this is expected with respect to centuries, since characters are unlikely to change the way they are written in time spans of consecutive years. As a result the KMeans algorithm, possibly merged 2 century classes into one.

## Part 5 | Unsupervised Clustering on BT1 Dataset
---

We will now cluster the second dataset of images where no ground truth is given. We will do so for every different character, computing the most optimal k for each different one, according to the Davies Bouldin criterion. In order to identify the optimal k, we will utilize the elbow criterion. We note that we will drop characters that have less than 10 image instances:

```python
ds2_vc = ds2['character'].tolist()
x_ds2 = ds2[ds2['character'].apply(lambda x: ds2_vc.count(x) >= 10)] # https://bit.ly/3ENZb6I | select category with more than x occurrences
```

The process we will follow is basically looping over all available dataset characters and for each one of them calculate the Davies Bouldin metric for values of k from 2 up to 10. According to these measures, we will then calculate the knee of the curve, which will lead us to the optimal k for that specific character:

```python
k_list = [x for x in range(2, 11)] # used for KneeLocator

acc = {}
ds2_characters = np.unique(x_ds2['character']).tolist()
for ch in ds2_characters: # loop over available characters
    curr_ds = x_ds2[x_ds2['character'] == ch] # select specific character instances
    curr_x_ds2 = np.array(curr_ds['flattend'].tolist()) # keep flattend images
    db_acc = []
    for k in range(2, 11): # compute optimal k
        kmeans = KMeans(n_clusters=k, random_state=39) # init KMeans
        pred = kmeans.fit_predict(curr_x_ds2) # predictions
        db_acc.append(davies_bouldin_score(curr_x_ds2, pred)) # DB index
    kl = KneeLocator(k_list, db_acc, curve="concave", direction="increasing") # calculate location of knee for all tested ks
    acc[ch] = (kl.knee, db_acc[kl.knee-2]) # save to dictionary

pd.DataFrame.from_dict(acc).rename(index={0: 'Optimal k', 1: 'Davies Bouldin Index'}) # pretty display
```

We will then visualize each cluster image for every character:

```python
figure(figsize=(18, 12))

i = 1 # subplot counter
index_acc = []
for key, value in acc.items():
    curr_ds = x_ds2[x_ds2['character'] == key] # fetch character ds
    curr_x_ds2 = np.array(curr_ds['flattend'].tolist()) # flattend for KMeans fitting
    kmeans = KMeans(n_clusters=value[0], random_state=39) # init KMeans with optimal k for character
    pred = kmeans.fit_predict(curr_x_ds2) # cluster of every image
    for cl in range(1, value[0]+1): # loop over different clusters of character
        x = np.argmax(pred==cl-1) # find first index of cluster
        ax = plt.subplot(5, 12, i)
        ax.set_xticks([])
        ax.set_yticks([])
        plt.imshow(curr_ds['image'].iloc[x], cmap='gray') # plot in grayscale
        i += 1

plt.show()
```

The different clusters that we calculated for every character, basically corresponds to different ways the characters is written. From the above images, we can see that `ω`, `χ` and `ι` are described with the most clusters among the dataset. In detail, we note the following cases:

- `α` indeed corresponds to 2 different written forms
- `ζ` is written in two significantly different ways, one of them looking very similar to the letter `L` from the latin alphabet
- character `θ` possibly enabled multiple clusters due to the accent we can see in the images
- we can notice artifacts in the images of `λ`, which possibly was the reason for dual clustering
- it seems like there is no obvious reason why `ι` was clustered in 5 different groups, while barely 2 can be explained
- `ξ` indeed corresponds to 2 different written forms
- the way `ο` is written is not exactly straightforward and is justifiably clustered into two separate groups
- it seems like there is no obvious reason why `χ` was clustered in 5 different groups
- it seems like there is no obvious reason why `ω` was clustered in 5 different groups
